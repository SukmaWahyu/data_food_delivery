```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:/food_delivery_datasets.csv')
```

```python
df.head()
```

```python

df['DTO'] = pd.to_datetime(df['date_time'])
df.head()
```

```python
df['DTO'] = pd.to_datetime(df['date_time'])

df['jam'] = df['DTO'].dt.hour
df.head()
```

```python
result=df.groupby(["jam"]).count()
result
```

```python
result = df.groupby(['jam'])['cust_id'].count()
hours = [jam for jam, df in df.groupby('jam')]

plt.plot(hours,result)
plt.xticks(hours)
plt.xlabel('jam')
plt.ylabel('frekuensi')
plt.grid()
plt.show()
```
